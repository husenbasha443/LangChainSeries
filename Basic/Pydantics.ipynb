## Simple Example

```python
from pydantic import BaseModel

class Person(BaseModel):
    name:str
    usn:int

```

```python
p1 = Person(name="Husen",usn=123)
print(p1)
```

```python
p2 = Person(name="Basha",usn="hgg")
print(p2)
```

## Using Optional Field

```python
from typing import Optional

class Employee(BaseModel):
    name:str
    ssn:int
    salary:int
    dept:Optional[str]='Finance'
    isWoked:Optional[bool]=True
```

## Using Optional Field

```python
e1 = Employee(name='Husen',ssn=452,salary=45000)
print(e1)
```

## Without Using Optional Field

```python
e2 = Employee(name='Basha',ssn=452,salary=8520,dept='IT',isWoked=False)
print(e2)
```

## Error Fields

```python
e3 = Employee(name='Maa',ssn='dgdg',salary=45200,)
##Validation Error
```

## Perform List

```python
from typing import List

class Mobile(BaseModel):
    brand:str
    ram:list[str]
    price:int
```

```python
m1 = Mobile(brand='Moto',ram=['4GB','8GB','16GB'],price=45000)
print(m1)
```

## Incorrect Formate

```python
m2 = Mobile(brand='Moto',ram=['4GB','8GB',12],price=45000)
print(m2)
```

## Nested Models

```python
class Student(BaseModel):
    name:str
    usn:int
    
class Course(BaseModel):
    cname:str
    cid:int
    student:Student
```

```python
c1 = Course(
    cname='aaaa',
    cid=456,
    student={'name': 'jjhh', 'usn': 2}
)
print(c1)
```
