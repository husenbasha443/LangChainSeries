## Pydantic Basics

```python
from pydantic import BaseModel
```

```python
class Person(BaseModel):
    name:str
    age:int
    city:str
```

```python
person = Person(name='Husen',age=23,city='Mudgal')
print(person)
type(person)
```

## USing DataClass

Using data classes data validation are not happen in Using DataClass

Instead of String datatype we can give Integer Valuesm

```python
from dataclasses import dataclass

@dataclass
class Person1:
    name:str
    age:int
    city:str
    
person = Person1(name='Husen',age=23,city=12)
print(person)
type(person)
    
    
```

## Model with Optional Fields

```python
from typing import Optional
from pydantic import BaseModel

class Student(BaseModel):
    name:str
    usn:int
    dept : Optional[str]='MBA'
```

## With OPtonal Values

```python
s1 = Student(name='Basha',usn=123)
print(s1)
type(s1)
```

## With Optional Values

```python
s2 = Student(name='Husen',usn=786,dept='M.Com')
print(s2)
```

# Using List

```python
class Employee(BaseModel):
    name:str
    ssn:int
    deptno:list[int]
```

```python
e1 = Employee(name='Husen',ssn=785,deptno=[45,52,45,25,78])
print(e1)
```

# Model with Nested Values

```python
class Shamid(BaseModel):
    name:str
    age:int
    
class Husen(BaseModel):
    hname:str
    shamid : Shamid
    
```

```python
husen = Husen(
    hname='Basha Sufi',
    shamid=Shamid(name='Husen', age=45)
)
print(husen)
```
