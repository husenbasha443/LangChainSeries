# HTML Header Splitter

```python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_string = """"

<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>HTML Tags Example</title>
</head>
<body>
    <h1>Welcome to HTML Tags Example</h1>
    <h2>Heading Tags</h2>
    <h3>This is H3</h3>
    <h4>This is H4</h4>

    <h5>This is H5</h5>
    <h6>This is H6</h6>

    <p>This is a <strong>paragraph</strong> with <em>italic</em> and <u>underlined</u> text.</p>

    <hr>
    <h1>Basha Sufi</h1>
    <h2>Image Tag</h2>
    <img src="https://via.placeholder.com/150" alt="Sample Image">

    <h2>Link Tag</h2>
    <a href="https://www.google.com" target="_blank">Go to Google</a>

    <h2>List Tags</h2>
    <h3>Ordered List</h3>
    <ol>
        <li>Item One</li>
        <li>Item Two</li>
        <li>Item Three</li>
    </ol>

    <h3>Unordered List</h3>
    <ul>
        <li>Apple</li>
        <li>Banana</li>
        <li>Cherry</li>
    </ul>

    <h2>Table Tag</h2>
    <table border="1" cellpadding="5">
        <tr>
            <th>Name</th>
            <th>Age</th>
        </tr>
        <tr>
            <td>John</td>
            <td>25</td>
        </tr>
        <tr>
            <td>Alice</td>
            <td>22</td>
        </tr>
    </table>

    <h2>Form Tag</h2>
    <form>
        <label>Name:</label>
        <input type="text" placeholder="Enter your name"><br><br>

        <label>Email:</label>
        <input type="email" placeholder="Enter your email"><br><br>

        <label>Message:</label><br>
        <textarea rows="4" cols="30"></textarea><br><br>

        <button type="submit">Submit</button>
    </form>

    <h2>Audio & Video Tags</h2>
    <audio controls>
        <source src="sample.mp3" type="audio/mpeg">
    </audio>

    <video width="300" controls>
        <source src="sample.mp4" type="video/mp4">
    </video>

    <h2>Div & Span Tags</h2>
    <div style="border:1px solid black; padding:10px;">
        This is a <span style="color:red;">sample span</span> inside a div.
    </div>

</body>
</html>

"""
 

header_split=[
    ('h1','Header-01'),
    ('h2','Header-02'),
    ('h3','Header-03')
]

html_split = HTMLHeaderTextSplitter(headers_to_split_on=header_split)
final_html_split=html_split.split_text(html_string)
final_html_split
```

# Complete Example

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Load The Text
speech=""
with open('Speech.txt') as f:
    speech=f.read()

speech   # Load The Text

#Split The Text
text_split = RecursiveCharacterTextSplitter(chunk_size=5,chunk_overlap=5)
final_split = text_split.create_documents([speech])
print(final_split) 

#from langchain_community.embeddings import OllamaEmbeddings
```

```python
from langchain_community.embeddings import OllamaEmbeddings

embeddings = (
    OllamaEmbeddings(model="gemma:2b")
)

documents = embeddings.embed_documents(final_split)
```

```python
# Convert Document objects → strings
texts = [doc.page_content for doc in final_split]

# Embeddings
embeddings = OllamaEmbeddings(model="gemma:2b")
vectors = embeddings.embed_documents(texts)

print(vectors)
```
