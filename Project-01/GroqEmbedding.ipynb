```python
from langchain_classic.document_loaders import PyPDFLoader

loader = PyPDFLoader('Rohan_SQL.pdf')

pdf_load = loader.load()
pdf_load
```

# PDF Splitter

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

split = RecursiveCharacterTextSplitter(chunk_size=100,chunk_overlap=50)
final_split = split.split_documents(pdf_load)
```
