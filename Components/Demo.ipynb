# Chat Prompt Template

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
    [
        ("system","What are key skills require AI Engineer. Provide an Answer Based on My Question"),
        ("user","{input}")
    ]
)
prompt
```

# Create a Chain

```python
chain = prompt|llm
response = chain.invoke({"input": 'Can You Tell me about key responsblity of AI Engineer'})
response

# | --> It Indicates the Sequential Exucution Process. It Exucute one by one process first it Exucute prompt and then it exucute the llm(call the model)
```
